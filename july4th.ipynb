```python
# Linear graph 
import matplotlib.pyplot as plt
plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
#Graph to plot distance and time
distance=list(range(2,11,2))
time=list(range(1,6))
distance
time
plt.title("Distance vs Time")#title
plt.plot(time,distance,'yo')#yellow bubbles due to this
plt.xlabel('Time')#xlabel name
plt.ylabel('Distance')#ylabel name
plt.grid()#for lines in graph
plt.show()
```

```python
# v=at
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
```

```python
# Numbers vs powers
n=list(range(10))
sqn=[i**2 for i in n]
plt.plot(n,sqn)
plt.show()
```

```python
#Voltage-Time
import numpy as np
t=[i/10 for i in range(1,21)]
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title('Time vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig("Images/Voltage-Time.png")# to store the result image in this location
plt.show()
```

```python
# plot Income of Arkansas from 2005 to 2013
import pandas as pd
df=pd.read_csv('Datafiles/Income.csv')
years=df.columns[2:]
incomeArkansas=df.values[3,2:]
plt.title('Income of Arkansas')
plt.xlabel('year')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# income of all states from 2005 to 2013
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
incomeallstates
plt.title('Income of all states')
plt.xlabel('years')
plt.ylabel('incomeallstates')
plt.plot(years,incomeallstates)
plt.show()
```
